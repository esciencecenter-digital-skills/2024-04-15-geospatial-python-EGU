```python
import hvplot.pandas
import osmnx as ox
```

```python
# get polygon of Rhodes
rhodes = ox.geocode_to_gdf("Rhodes")
rhodes.hvplot(tiles="ESRI")
```

```python
rhodes.to_file("rhodes.gpkg")
```

```python
# get highways
highways = ox.features_from_place(
    "Rhodes",
    tags={
        "highway":[
            "primary", 
            "secondary", 
            "tertiary",
        ]
    }
)
highways.hvplot(tiles="ESRI")
```

```python
epsg_raster = 32635  # get EPSG from raster data
highways_repr = highways.to_crs(epsg_raster)
```
